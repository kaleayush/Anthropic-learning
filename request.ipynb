```python
#Step create an API client 
from anthropic import Anthropic
import os
client = Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
#model = "claude-sonnet-4-0" 
```

```python
# Make a request
message= client.messages.create(
    model = "claude-sonnet-4-5" ,
    max_tokens= 1000,
    messages=[
        {
            "role": "user",
            "content": "what is quantum computing? Answer in one sentence"
        }
    ]
)
print(message)
# in this No response will be generated because I have not purchased api key form anthropic 
```

```python
import types

from google import genai
from google.genai import types  # this is where 'types' comes from
# gemini-2.0-flash       Fast, efficient (recommended)
# gemini-2.5-pro         Most capable
# gemini-2.0-flash-lite  Cheapest/fastest

client = genai.Client(api_key='GEMINI_API_KEY')

# response = client.models.generate_content(
#     model="gemini-2.5-flash",
#     contents="what is quantum computing? Answer in one sentence"
# )
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[types.Part.from_text(text="Explain what is quantum computing with example")],
    config=types.GenerateContentConfig(
    max_output_tokens=1000,
    temperature=0.7,       # creativity (0 = deterministic, 1 = creative)
    top_p=0.9,
    )
)
print(response.text)

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[types.Part.from_text(text="write another sentence")],
    config=types.GenerateContentConfig(
    max_output_tokens=1000,
    temperature=0.7,       # creativity (0 = deterministic, 1 = creative)
    top_p=0.9,
    )
)
# response = client.models.generate_content(
#     model="gemini-2.0-flash",
#     contents=[
#              types.Part.from_text(text='What is shown in this image?'),

#     ],
#     config=types.GenerateContentConfig(
#         max_output_tokens=1000,
#     )
# )
print(response.text)
```
